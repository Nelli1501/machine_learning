```python
import numpy as np
import matplotlib.pyplot as plt
```

### Функция одной переменной

```python
x0 = 5
l = 0.1
res = []
```

```python
def f_x(x):
    f = 3*x**2
    return f
```

```python
def df_x(x):
    df = 6*x
    return df
```

```python
def gradient_descent(x0, l):
    res = []
    while True:
        res.append([x0, f_x(x0)])
        x1 = x0 - l*df_x(x0)
        if np.linalg.norm(x0-x1) <= 0.00000001:
            break
        x0=x1
    return np.array(res)
```

```python
res_new = gradient_descent(5, 0.1)
```

```python
x = np.linspace(-5, 5, 100)
y = f_x(x)

res_x = res_new[:, 0]
res_y = res_new[:, 1]

plt.figure()
plt.plot(x, y, label='3x^2')
plt.plot(res_x, res_y, marker='o', color='red')
```

#### Метод наискорейшего спуска

```python
from scipy.optimize import minimize

def g(l):
    x1 = x0 - l*df_x(x0)
    return f_x(x1)

l = 0 
result = minimize(g, l)

optimal = result.x[0] 
res_optimal = gradient_descent(x0, optimal)

x = np.linspace(-5, 5, 100)
y = f_x(x)

res_x = res_optimal[:, 0]
res_y = res_optimal[:, 1]

plt.figure()
plt.plot(x, y, label='3x^2')
plt.plot(res_x, res_y, marker='o', color='red')
```

### Функция двух переменных

```python
def f(x, y):
    f1 = 3*x**2 + 5*y**2
    return f1
def df(x, y):
    df1 = 6*x
    df2 = 10*y
    return df1, df2
```

```python
x0 = 5
y0 = 10
def gradient_descent(x0, y0, l):
    res1 = []
    while True:
        res1.append([x0, y0, f(x0, y0)])
        df1, df2 = df(x0, y0)
        x1 = x0 - l*df1
        y1 = y0 - l*df2
        if np.linalg.norm(np.array([x1-x0])-np.array([y1-y0])) <= 0.00000001:
            break
        x0 = x1
        y0 = y1
    return np.array(res1)
```

```python
res_new = gradient_descent(5, 10, 0.1)
res_new
```

```python
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)

res_x = res_new[:, 0]
res_y = res_new[:, 1]
res_z = res_new[:, 2]


ax.plot(res_x, res_y, res_z, marker='o', color='r')

plt.show()
```
