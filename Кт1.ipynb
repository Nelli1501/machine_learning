```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set();

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('city_united.csv')
df
```

```python
df.info() #пропусков нет
```

```python
#Создаем x и y
x = df.drop(['UA_Name', 'Quality of Life Index'], axis=1)
y = df['Quality of Life Index']
```

```python
#Сделаем обработку категориальных переменных
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
x['UA_Country'] = encoder.fit_transform(x['UA_Country'])
x['UA_Continent'] = encoder.fit_transform(x['UA_Continent'])
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
import math 
from sklearn.metrics import r2_score
```

```python
#Построим регрессию исходной модели
reg = LinearRegression().fit(train_x, train_y)
predictions = reg.predict(test_x)
Ri = r2_score(test_y, predictions)
MSE = mean_squared_error(test_y, predictions)
print("Исходная модель: Rквадрат:", Ri, "MSE:", MSE)
```

```python
#Отберем признаки на основе пермутационной важности
import eli5
from eli5.sklearn import PermutationImportance
```

```python
perm = PermutationImportance(reg, random_state=42).fit(test_x, test_y)
```

```python
eli5.show_weights(perm, feature_names=list(test_x.columns))
```

```python
column1 = ['Venture Capital', 'Business Freedom', 'Education',
       'Environmental Quality', 'Economy']
```

```python
train_x_new = train_x[column1]
test_x_new = test_x[column1]
reg = LinearRegression().fit(train_x_new, train_y)
predictions = reg.predict(test_x_new)
Ri = r2_score(test_y, predictions)
MSE = mean_squared_error(test_y, predictions)
print("Новая модель: Rквадрат:", Ri, "MSE:", MSE)
```

Модель стала лучше

```python
#Отберем признаки на основе f-статистики
from sklearn.feature_selection import f_regression, SelectKBest
f_statistic, p_values = f_regression(train_x, train_y)
sekector = SelectKBest(f_regression, k=5)
sekector.fit_transform(train_x, train_y)
```

```python
column2 = sekector.get_feature_names_out()
column2
```

```python
train_x_new = train_x[column2]
test_x_new = test_x[column2]
reg = LinearRegression().fit(train_x_new, train_y)
predictions = reg.predict(test_x_new)
Ri = r2_score(test_y, predictions)
MSE = mean_squared_error(test_y, predictions)
print("Новая модель: Rквадрат:", Ri, "MSE:", MSE)
```

Модель стала лучше, то отличается (в худшую сторону) в отличие от первого

```python
#Сделаем полиномиальные признаки
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(2)
train_X_poly = pd.DataFrame(poly.fit_transform(train_x), columns=poly.get_feature_names_out())
test_X_poly = pd.DataFrame(poly.fit_transform(test_x), columns=poly.get_feature_names_out())
```

```python
reg = LinearRegression().fit(train_X_poly, train_y)
predictions = reg.predict(test_X_poly)
Ri = r2_score(test_y, predictions)
MSE = mean_squared_error(test_y, predictions)
print("Исходная модель: Rквадрат:", Ri, "MSE:", MSE)
```

```python
from sklearn.feature_selection import f_regression, SelectKBest
f_statistic, p_values = f_regression(train_X_poly, train_y)
sekector = SelectKBest(f_regression, k=20)
sekector.fit_transform(train_X_poly, train_y)
```

```python
column3 = sekector.get_feature_names_out()
column3
```

```python
train_X_new = train_X_poly[column3]
test_X_new = test_X_poly[column3]
reg = LinearRegression().fit(train_X_new, train_y)
predictions = reg.predict(test_X_new)
Ri = r2_score(test_y, predictions)
MSE = mean_squared_error(test_y, predictions)
print("Новая модель: Rквадрат:", Ri, "MSE:", MSE)
```
